```python
# Import dependencies
import pandas as pd
import os
import datetime as dt
```

```python
# Get CWD
cwd = os.getcwd()
cwd
```

```python
# View each filenames month - Filename format is 'YYYY-MM_JC.csv'
for file in os.listdir(cwd + "\\Resources\\"):
    filename = os.fsdecode(file)
    if filename.endswith(".csv"):
        print(filename.split("-")[1].split("_")[0])
```

```python
# Create two lists to concat the dfs

# One list for winter months
# One list for summer months
winter_dfs = []
summer_dfs = []

# Loop through each file in the Resources directory
for file in os.listdir(cwd + "\\Resources\\"):
    filename = os.fsdecode(file)
    if filename.endswith(".csv"):
        # Check the month of the file name - If it's a winter month, append to the 'winter_dfs' list
        if (int(filename.split("-")[1].split("_")[0]) == 11 or int(filename.split("-")[1].split("_")[0]) == 12 or int(filename.split("-")[1].split("_")[0]) == 1):
            df = pd.read_csv(cwd+"\\Resources\\"+filename, dtype={'start_station_id':"string", "end_station_id":"string"})
            winter_dfs.append(df)
        else:
            # If its not a winter month - append to the 'summer_dfs' list 
            df = pd.read_csv(cwd+"\\Resources\\"+filename, dtype={'start_station_id':"string", "end_station_id":"string"})
            summer_dfs.append(df)
    else:
        # If the file is not a csv, continue to the next file
        continue
```

```python
# Concat the summer months into one df
summer_df = pd.concat(summer_dfs)

# Reset the index to remove duplicate values
summer_df = summer_df.reset_index(drop=True)

# Preview the df
summer_df.head()
```

```python
# Concat the winter months into one df
winter_df = pd.concat(winter_dfs)

# Reset the index to remove duplicate values
winter_df = winter_df.reset_index(drop=True)

# Preview the df
winter_df.head()
```

```python
# Convert start and end date/time to correct dtype
summer_df[["started_at","ended_at"]] = summer_df[["started_at","ended_at"]].astype("datetime64[ns]", errors="raise")
winter_df[["started_at","ended_at"]] = winter_df[["started_at","ended_at"]].astype("datetime64[ns]", errors="raise")

# Check the dtypes are correct
print(summer_df.info())
print("-" * 50)
print(winter_df.info())
```

```python
# Add a column to join the weather data to in Tableau
summer_df['weather_join'] = summer_df['started_at'].dt.strftime("%Y-%m-%d %H:00:00")
winter_df['weather_join'] = winter_df['started_at'].dt.strftime("%Y-%m-%d %H:00:00")
```
